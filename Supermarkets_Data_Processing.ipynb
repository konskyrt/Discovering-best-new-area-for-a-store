```python
import numpy as np
import pandas as pd
```

```python
supermarkets = pd.read_csv('Zurich_supermarkets2.csv', engine='python')
#supermarkets['PLZ']=[None]
supermarkets = supermarkets.drop_duplicates()
supermarkets
```

```python
# get PLZ (postcode)into a new column. Help from Cassandra

def PLZ_extract(address):
    
    try:
        comma_position = address.find(",")
        if comma_position == -1:
            PLZ = address[0:4]
        else: 
            PLZ = address[comma_position+2:comma_position+6]

        if PLZ.isdigit() == False:
            address = address[comma_position+1:]
            PLZ = PLZ_extract(address)
            return PLZ
        else: 
            return PLZ
    except Exception:
        PLZ = "somethin went wrong: " + str(Exception)
        return PLZ
supermarkets["PLZ"] = supermarkets['Adress'].apply(lambda x: PLZ_extract(x))    

supermarkets
```

```python
print(supermarkets[supermarkets["PLZ"] == "somethin went wrong: <class 'Exception'>"])
```

```python
# decide if a supermarket is Migros as names have some variability
def check_migros(name):
    if name.lower().find("migro")==-1:
        return 0
    else:
        return 1
    
supermarkets["Migros"] = supermarkets['Name'].apply(lambda x: check_migros(x))    

supermarkets
```

```python
# pivot by PLZ and count migros supermarkets
migros_count = supermarkets.groupby('PLZ').agg(num_migros = ('Migros', 'sum'))
migros_count

# pivot by PLZ and count all supermarkets
all_supermarket_count= supermarkets.groupby('PLZ').agg(num_all = ('Migros', 'count'))
all_supermarket_count

# merge two tables
supermarket_count = pd.merge(all_supermarket_count, migros_count, left_on = 'PLZ', right_on = 'PLZ')

#PLZ_subset = ["8001", "8003", "8004", "8005", "8008", "8021", "8032", "8037", 
#              "8045", "8046", "8048", "8050", "8051", "8052", "8055", "8057"]
#PLZ_subset  = pd.DataFrame(PLZ_subset)
#PLZ_subset.columns = ["PLZ"]
#print(PLZ_subset)

# Couldn't subset, so used merge
# supermarket_count = supermarket_count[supermarket_count['PLZ'] == PLZ_subset]
#supermarket_count = pd.merge(PLZ_subset, supermarket_count, left_on = 'PLZ', right_on = 'PLZ')

# supermarket_count.set_index("PLZ")
supermarket_count
```
